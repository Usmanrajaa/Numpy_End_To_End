# 6. NumPy Searching, Sorting, and Filtering

```python
arr = np.array([1, 3, 5, 7, 9])

# Find indices where values > 4
indices = np.where(arr > 4)
print(indices)# (array([2, 3, 4]),)
```

```python
# Find exact matches
indices = np.where(arr == 5)# (array([2]),)# Find non-matches
indices = np.where(arr != 5)
```

```python
sorted_arr = np.array([10, 20, 30, 40, 50])

# Find where 25 would be inserted
pos = np.searchsorted(sorted_arr, 25)# 2
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])

# Multiple conditions (AND)
indices = np.where((arr > 2) & (arr < 5))# (array([2, 3]),)# Multiple conditions (OR)
indices1 = np.where((arr < 2) | (arr > 5))# (array([0, 5]),)
```

```python
indices1
```

```python
# Return new sorted array
sorted_arr = np.sort(arr)# [1, 2, 3, 4, 5]# In-place sorting
arr.sort()# Modifies original array
```

```python
arr_2d = np.array([[3, 1], [5, 2], [4, 0]])

# Sort each row
row_sorted = np.sort(arr_2d, axis=1)

# Sort each column
col_sorted = np.sort(arr_2d, axis=0)

# Sort by specific column (fancy indexing)
sorted_by_col1 = arr_2d[arr_2d[:,1].argsort()]
```

```python
arr = np.array([10, 20, 30, 40, 50])

# Simple filter
filtered = arr[arr > 25]# [30, 40, 50]# Multiple conditions
filtered = arr[(arr > 15) & (arr < 45)]# [20, 30, 40]
```

```python
condition = arr > 25
filtered = np.extract(condition, arr)# [30, 40, 50]
```

```python
mask = np.array([True, False, True, False, True])
filtered = arr[mask]# [10, 30, 50]
```

```python
filtered
```

```python
# Remove outliers
data = np.array([1, 2, 3, 100, 2, 3])
cleaned = data[(data > 0) & (data < 10)]
```
